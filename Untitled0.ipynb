```python
def data_quality_check(df):
    quality_report = {
        'total_records': len(df),
        'duplicate_rows': df.duplicated().sum(),
        'missing_values': df.isnull().sum().to_dict(),

        # Invalid numeric checks
        'invalid_sales': pd.to_numeric(df['sales'], errors='coerce').isnull().sum(),
        'invalid_quantity': (df['quantity'] < 0).sum(),
        'invalid_discount': ((df['discount'] < 0) | (df['discount'] > 1)).sum(),
        'invalid_profit': df['profit'].isnull().sum(),
        'invalid_shipping_cost': (df['shipping_cost'] < 0).sum(),

        # Invalid year check
        'invalid_year': ((df['year'] < 2000) | (df['year'] > 2030)).sum(),

        # Empty string checks for object columns
        'empty_ship_mode': (df['ship_mode'].str.strip() == '').sum(),
        'empty_segment': (df['segment'].str.strip() == '').sum(),
        'empty_state': (df['state'].str.strip() == '').sum(),
        'empty_product_name': (df['product_name'].str.strip() == '').sum()
    }

    return quality_report
```

```python
quality_report = data_quality_check(df)
print(quality_report)
```

```python
# For categorical columns
df_clean['ship_mode'] = df_clean['ship_mode'].fillna('Unknown')
df_clean['segment'] = df_clean['segment'].fillna('Unknown')
df_clean['state'] = df_clean['state'].fillna('Unknown')
df_clean['country'] = df_clean['country'].fillna('Unknown')
df_clean['market'] = df_clean['market'].fillna('Unknown')
df_clean['region'] = df_clean['region'].fillna('Unknown')
df_clean['category'] = df_clean['category'].fillna('Unknown')
df_clean['sub_category'] = df_clean['sub_category'].fillna('Unknown')
df_clean['product_name'] = df_clean['product_name'].fillna('Unknown')
df_clean['product_id'] = df_clean['product_id'].fillna('Unknown')

# For numeric columns
df_clean['sales'] = pd.to_numeric(df_clean['sales'], errors='coerce')
df_clean['sales'] = df_clean['sales'].fillna(0)

df_clean['quantity'] = df_clean['quantity'].fillna(0)
df_clean['discount'] = df_clean['discount'].fillna(0)
df_clean['profit'] = df_clean['profit'].fillna(0)
df_clean['shipping_cost'] = df_clean['shipping_cost'].fillna(0)
df_clean['year'] = df_clean['year'].fillna(0)

# Summary of remaining missing values
missing_summary = df_clean.isnull().sum()
print(f"Remaining nulls:\n{missing_summary}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
```

```python
svc = SVC(kernel='sigmoid', gamma=1.0)
knc = KNeighborsClassifier()
mnb = MultinomialNB()
dtc = DecisionTreeClassifier(max_depth=5)
lrc = LogisticRegression(solver='liblinear', penalty='11')
rfc = RandomForestClassifier(n_estimators=50, random_state=2)
abc = AdaBoostClassifier(n_estimators=50, random_state=2)
bc = BaggingClassifier(n_estimators=50, random_state=2)
etc = ExtraTreesClassifier(n_estimators=50, random_state=2)
gbdt = GradientBoostingClassifier(n_estimators=50, random_state=2)
xgb = XGBClassifier(n_estimators=50, random_state=2)
```

```python
clfs = {
    'SVC' : svc,
    'KN' : knc,
    'NB' : mnb,
    'DT' : dtc,
    'LR' : lrc,
    'RF' : rfc,
    'BGL' : bc,
    'ETC' : etc,
    'GBDT' : gbdt,
    'AdaBoost' : abc,
    'xgb' : xgb

}
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score

def train_classifier(clf, X_train, y_train, X_test, y_test):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)

    return accuracy, precision
```

```python
print("For" )
```

```python
accuracy_list = []
precision_list = []

for name, clf in clfs.items():
    try:
        current_accuracy, current_precision = train_classifier(
            clf, X_train, y_train, X_test, y_test
        )

        print("For", name)
        print("Accuracy -", current_accuracy)
        print("Precision -", current_precision)

        accuracy_list.append(current_accuracy)
        precision_list.append(current_precision)

    except Exception as e:
        print(f"Error in {name}: {e}")
```
